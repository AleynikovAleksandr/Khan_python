```python
from google.colab import drive
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "/content/drive/My Drive/100_Sales_Records.xlsx"
df = pd.read_excel(file_path, sheet_name="sheet 1")
```

```python
df_priority_online = df[(df["Order Priority"].isin(["H", "C"])) & (df["Sales Channel"] == "Online")]
display(df_priority_online.head())
```

```python
df_filtered = df[df["Total Profit"] > 500000]
display(df_filtered.head(10))
```

```python
df_sorted = df_filtered.sort_values(by="Total Profit", ascending=False).drop_duplicates().head(10)
display(df_sorted)
```

```python
df_grouped = df.groupby("Item Type")["Total Profit"].mean().reset_index()
display(df_grouped)
```

```python
df_region_stats = df.groupby("Region").agg({
    "Units Sold": "sum",
    "Total Profit": "mean",
    "Unit Price": "mean"
}).reset_index()
display(df_region_stats)
```

```python
missing_values = df.isnull().sum().sum()
print(f"Общее количество пропущенных значений: {missing_values}")
```

```python
pivot_profit = df.pivot_table(
    values="Total Profit",
    index="Region",
    columns="Order Priority",
    aggfunc="mean"
).round(2).fillna(0)
display(pivot_profit)
```

```python
pivot_summary = df.pivot_table(
    values=["Total Revenue", "Total Profit"],
    index="Region",
    aggfunc="sum"
).round(2)

pivot_summary["Profit Margin (%)"] = (
    pivot_summary["Total Profit"] / pivot_summary["Total Revenue"] * 100
).round(2)

display(pivot_summary)
```

```python
sales_channel_counts = df["Sales Channel"].value_counts()

plt.style.use("dark_background")

online_sales = sales_channel_counts.get("Online", 0)
offline_sales = sales_channel_counts.get("Offline", 0)

plt.figure(figsize=(8, 8))
plt.pie(
    [online_sales, offline_sales],
    labels=["Online", "Offline"],
    autopct="%1.1f%%",
    colors=["skyblue", "lightcoral"]
)
plt.title("Доли каналов продаж (Онлайн vs Оффлайн)")
plt.show()
```

```python
profit_by_country = df.groupby("Country")["Total Profit"].sum()

plt.figure(figsize=(10, 6))
profit_by_country.head(20).plot(kind='bar', color='#008B8B', edgecolor='black')

plt.gca().set_facecolor('black')
plt.grid(True, which='both', color='white', linestyle='-', linewidth=0.5)

plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')

plt.xlabel('Country', color='white')
plt.ylabel('Total Profit', color='white')
plt.title('Топ-20 стран по общей прибыли', color='white')

plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.to_period("M")
monthly_profit = df.groupby("Month")["Total Profit"].sum()

plt.figure(figsize=(16, 8))
plt.plot(monthly_profit.index.astype(str), monthly_profit.values, marker="o", color="dodgerblue", linestyle="-", linewidth=2)

plt.ylabel("Общая прибыль ($)", fontsize=14, color="white")
plt.title("Динамика прибыли по месяцам", fontsize=16, color="white")
plt.grid(color="white", linestyle="--", alpha=0.5)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
